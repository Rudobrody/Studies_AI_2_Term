# 0.0 Theory about superresolution 

Low resolution pixel could have been generated from an infinite number of different high resolution pixel configuration. Therefore, the model must learn a plausible mapping from the low resolution space to the high resolution space (textures, edges, fine patterns) that were lost during the downscaling process.

What is advantage of self-supervised method is that we can generate our training data automatically, without human interventon like manually labeling. 

# Examples of architecture

- Efficient Post-Unsampling architectures (ESPCN, FSRCNN), LR image is fed directly into a deep CNN, network extracts complex features while maintaining the small spatial dimensions. At the end of the network there was a special unsmapling layer, known as the PixelShuffle or Sub-Pixel Convolution, rearranges the feature maps to produce the final HR image.
- Residual learning - instead of trying to learn a direct mapping from LR to HR, the network is tasked with learning the residual image, so the difference betweeen the HR and LR images. The network output is then simply added back to an upscaled version if the input to form the final HR image.   
- Generative Adversarial Networks, the generator is the SR network itself (based on a residual architecture). It takes an LR image and generates a candidate HR image. The discriminator - this is a classifier network. It is trained to distinguish real HR images and 'fake' HR images produced by the Generaotr. Generator is rewarded for creating images that the Discriminator classifies as 'real'. It gives photorealistic results. 


# Loss function

- Pixel-Wise Loss (L1, l2/mse) - difference between the pixels, tends to produce blurry results by averging possible solutions
- Perceptual loss - Instead of comparing pixels, we compare high-level features. We feed both the generated image and the ground truth image into a pre-trained classification network (VGG19) and calculate the difference btween their feature representation at intermediate layers. 
- Adversarial loss - loss from the GAN setup. It measures how well the generator is fooling discriminator. 

Modern GAN-based SR models use a weighted combination of all three losses.

# Pipeline

1. Start with a large dataset of high-quality, high-resolution images  
2. For each high resoltuion image, we create a corresponding low resolution version applying a downscaling algorithm (like bcubic downsampling)  
3. Now we have a pair of images: low and high resolution 
4. The neural network is trained to take the low resoultion images as input and produce an output that is as close as possible to the original high resolution image.
